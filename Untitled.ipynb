##### Data Preprocessing

```python
df["Churn"] = np.where(df["Churn"] == "Yes",1,0)
```

```python
x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
```

```python
def preprocessing(data):
    scaler = StandardScaler()
    #numeric_cols = x.select_dtypes(include=["int","float"])
    numeric_cols = ["tenure","MonthlyCharges","TotalCharges"]
    data[numeric_cols] = scaler.fit_transform(data[numeric_cols])
    
    cat_cols = data.select_dtypes(include=["object"]).columns
    data = pd.get_dummies(data,cat_cols,drop_first=True)
    return data
```

```python
processed_df = preprocessing(x)
```

##### Feature Selection

```python
rf = RandomForestClassifier()
rf.fit(processed_df,y)
result_1 = pd.DataFrame({"feature":processed_df.columns,"RF_Importance":rf.feature_importances_}).sort_values(by="RF_Importance",ascending=False)
result_1
```

```python
xg = xgb.XGBClassifier()
xg.fit(processed_df,y)
result_2 = pd.DataFrame({"feature":processed_df.columns,"XG_Importance":xg.feature_importances_}).sort_values(by="XG_Importance",ascending=False)
result_2
```

```python
gb = GradientBoostingClassifier()
gb.fit(processed_df,y)
result_3 = pd.DataFrame({"feature":processed_df.columns,"GB_Importance":gb.feature_importances_}).sort_values(by="GB_Importance",ascending=False)
result_3
```

```python
# Step 1: Merge the DataFrames on the 'feature' column
result_1.set_index("feature", inplace=True)
result_2.set_index("feature", inplace=True)
result_3.set_index("feature", inplace=True)

# Concatenate along columns (axis=1)
merged_df = pd.concat([result_1, result_2, result_3], axis=1).reset_index()

# Step 2: Create a pivot table
pivot_table = pd.pivot_table(merged_df, index="feature")
```

By looking at feature importance value we can say that StreamingMovies_No internet service is least important column

```python
processed_df.drop("StreamingMovies_No internet service",axis=1,inplace=True)
```

##### Splitting Treatmnet and Imbalance Treatment

```python
x_train,x_test,y_train,y_test = train_test_split(processed_df,y,random_state=101,test_size=0.2,stratify=y)
```

```python
smt = SMOTE()
x_smote,y_smote = smt.fit_resample(x_train,y_train)
```

```python

# Define KFold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

best_model = None
best_score = 0

for train_index, test_index in kf.split(x_smote):
    X_train, X_val = x_smote.iloc[train_index], x_smote.iloc[test_index]
    y_train, y_val = y_smote.iloc[train_index], y_smote.iloc[test_index]
    
    # Initialize and train the model
    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)
    
    # Evaluate model performance on the validation fold
    y_pred = model.predict(X_val)
    score = accuracy_score(y_val, y_pred)
    print(f"Fold accuracy: {score}")
    
    # Update best model if current model is better
    if score > best_score:
        best_score = score
        best_model = model

print(f"Best fold accuracy: {best_score}")

# Get predictions from the best model on the test set
y_test_pred = best_model.predict(x_test)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Test set accuracy of the best model:", test_accuracy)
```
